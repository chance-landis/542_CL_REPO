# Finding the avg road surface temp over a week period
## Time Period: 2025 Apr 16 12AM - 2025 Apr 23 11:45PM

### Data was aggregated and filtered through the portals built in data features

```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Road_Weather_Information_Stations_20250424.csv')

# Rename the columns for readability
df = df.rename(columns={
    'StationName': 'Station Name',
    'RoadSurfaceTemperature': 'Road Surface Temperature'
})

# Sort data for better visualization
df_sorted = df.sort_values(by='Road Surface Temperature', ascending=False)

# Determine min and max for zooming in with a slight margin
min_temp = df_sorted['Road Surface Temperature'].min() - 0.5
max_temp = df_sorted['Road Surface Temperature'].max() + 0.5

# Create the plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.barh(df_sorted['Station Name'], df_sorted['Road Surface Temperature'], color='skyblue')
plt.xlabel('Average Road Surface Temperature (Â°F)')
plt.ylabel('Road Station')
plt.title('Average Road Surface Temperature by Station')
plt.xlim(min_temp, max_temp)  # Zoom in
plt.gca().invert_yaxis()  # Highest temp at the top
plt.tight_layout()
plt.show()
```
